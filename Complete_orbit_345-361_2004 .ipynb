```python
ncp_list = []
ave_ncp = []
std_ncp = []
for value in ncps.values():
    ncp_list.append(value)
print(len(ncp_list[0]))
for i in range(14):
    ncp_S = ncp_list[i]
    ave_value = np.sum(ncp_S, axis =0)/len(ncp_S)
    ave_ncp.append(ave_value)
    std = np.sqrt(np.sum((ncp_S - ave_value)**2, axis = 0)/len(ncp_S))
    std_ncp.append(std)
    
```
