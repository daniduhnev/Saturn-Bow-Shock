```python
ncps = {}
shock_vs = {}
aves = {}
stds = {}
cross_no = 1
for x in cross:
    for i in range(2):
        ncp,shock_value,ave,std = ncp_shock4(x[2*i],x[1],5,cross_no)
        ncps['ncp'+str(cross_no)] = ncp
        shock_vs['shock_value'+str(cross_no)] = shock_value
        aves['average_shock'+str(cross_no)] = ave
        stds['std'+str(cross_no)] = std
        cross_no += 1
```
